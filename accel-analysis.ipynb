```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

accel = pd.read_csv('./data/accel.csv')

freq = 20
dat = accel[['x','y','z']].values

def normalize(v):
    mag = np.linalg.norm(v)
    return v / mag

# determine orientation
g = normalize(np.mean(dat, axis=0))
print('g = {}'.format(g))

# multiply by gravity vector to get vertical acceleration
dat1 = np.dot(dat, g)

# plot the spectrogram
plt.figure(figsize=(18,10))
plt.specgram(dat1, Fs=freq, NFFT=32, noverlap=16);
```

Looking at the spectrogram, the steps seem to be 1-4 Hz.

```python
lower = 1 #Hz
upper = 3 #Hz
nyq = freq/2

# calculate coefficients and initial state
b, a = signal.butter(4, [lower/nyq, upper/nyq], btype='band')
zi = (signal.lfilter_zi(b, a))

print('''b = {}
a = {}
zi = {}'''.format(b, a, zi))

# apply filter
dat2, zf = signal.lfilter(b, a, dat1, zi=zi * dat1[0])

# plot the signal before and after filtering
plt.figure(figsize=(18,10))
plt.plot(dat2);
plt.plot(dat1);

# calculate and mark threshold
min = np.min(dat2)
max = np.max(dat2)
rms = np.sqrt(np.mean(dat2**2))
print('''min = {}
max = {}
rms = {}'''.format(min, max, rms))

threshold = rms / 2

plt.axhline(min, c='black')
plt.axhline(max, c='black')
plt.axhline(-threshold, c='red')
plt.axhline(threshold, c='red')
```

A 4th order Butterworth filter with a passband of 1-3 Hz seems to work pretty well. You can see the alternating steps in the filtered signal above, but most of the noise and harmonics are removed.

Using a detection threshold of rms/2 will detect the steps pretty well on this filtered signal.

```python
# simple threshold based counting function
def count_steps(dat, upper, lower):
    cnt = 0
    dir = 1
    for x in dat:
        if dir > 0:
            if x > upper:
                dir = -1
        elif dir < 0:
            if x < lower:
                dir = 1
                cnt = cnt + 1
    return cnt

steps = count_steps(dat2, threshold, -threshold)

print('{} steps'.format(steps))
```
